```python
outFile = open('Files_Directory/Twitter/twitterResults.txt','r')
#since some tweets are empty(\n), use while loop
tweets = []
num_tweets = 0
while num_tweets < 50:
    tweet = outFile.readline()
    if tweet != '\n':
        tweets.append(json.loads(tweet)['text'])
        num_tweets += 1
outFile.close()
```

```python
### Previous way of implementing the API call, one tweet at a time -

import requests
azure_keyphrases = []
azure_sentiments = []

azid=1

for tweet in tweets:
    
    
    # Azure portal URL.
    #Use the endpoint given by Azure as the base url.

    base_url = 'https://mgta414-wenhsuanhung.cognitiveservices.azure.com/'
    # The unique account key goes here.
    account_key = 'cbe1a5a393094a609d318924e70f6519'
        
    headers = {'Content-Type':'application/json', 'Ocp-Apim-Subscription-Key':account_key}

    documents = {'documents' : [
      {'id': azid, 'language': 'en', 'text': tweet}]}

    # Detect key phrases.
    batch_keyphrase_url = base_url + 'text/analytics/v3.1/keyPhrases'
    headers   = {'Ocp-Apim-Subscription-Key': account_key}
    response  = requests.post(batch_keyphrase_url, headers=headers, json=documents)
    key_phrases = response.json()
#     print(key_phrases)

    azure_keyphrases.append(key_phrases['documents'][0]['keyPhrases'])
    
    #Detect Sentiment
   
    batch_keyphrase_url = base_url + 'text/analytics/v3.1-preview.2/sentiment'
    headers   = {'Ocp-Apim-Subscription-Key': account_key}
    response  = requests.post(batch_keyphrase_url, headers=headers, json=documents)
    sentiment = response.json()
    
    azure_sentiments.append(sentiment['documents'][0]['sentiment'])
    azid+=1
```

## average sentiment

```python
negative = azure_sentiments.count('negative')
positive = azure_sentiments.count('positive')
neutral = azure_sentiments.count('neutral')
print(negative)
print(positive)
print(neutral)

total = negative + positive + neutral
```

```python
neutral_proportion = neutral / total
negative_proportion = negative / total
positive_proportion = positive / total

print('negative_proportion =' + str(negative_proportion))
print('positive_proportion =' + str(positive_proportion))
print('neutral_proportion =' +  str(neutral_proportion))
```

```python
sentiment_dict = {'neutral': neutral_proportion, 'positive': positive_proportion, 'negative': negative_proportion}
avg_sentiment = max(sentiment_dict, key=sentiment_dict.get)
print('for ' + input_word + ', the average sentiment is ' + avg_sentiment )
```
